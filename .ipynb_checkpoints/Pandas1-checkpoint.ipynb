# Pandas

```python
# Importing pandas
import numpy as np
import pandas as pd
```

```python
print("Pandas version:", pd.__version__)
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
# Creating series (array with labels)
series = pd.Series([2,3,7,11,13,17,19,23])
print(series)
```

```python
# creating dataframe from series
series_df = pd.DataFrame({
    'A': range(1,5), # returns 4 elements
    'B': pd.Timestamp('20200728'),
    'C': pd.Series(5,index=list(range(4)), dtype='float64'),
    'D': np.array([3]*4, dtype='int64'),
    'E': pd.Categorical(["Depression", "Social Anxiety", "Bipolar Disorder", "Eating Disorder"]),
    'F': 'Mental health',
    'G': 'is challenging'
})
print(series_df)
```

```python
# Creating dataframe from Dictionary
dict_df = [{'A': 'Apple', 'B': 'Ball'},
           {'A': 'Aeroplane', 'B': 'Bat', 'C': 'Cat'}]
dict_df = pd.DataFrame(dict_df)
print(dict_df)
```

```python
# Creating a dataframe from ndarrays
sdf = {
    'County':['Østfold', 'Hordaland', 'Oslo', 'Hedmark', 'Oppland', 'Buskerud'],
    'ISO-Code':[1,2,3,4,5,6],
    'Area': [4180.69, 4917.94, 454.07, 27397.76, 25192.10, 14910.94],
    'Administrative centre': ["Sarpsborg", "Oslo", "City of Oslo", "Hamar", "Lillehammer", "Drammen"]
    }
sdf = pd.DataFrame(sdf)
print(sdf)
```

## Combining numpy and Pandas

```python
np.random.seed(2)
array = np.linspace(1,10,10)

df = pd.DataFrame({'F': array})

df = pd.concat([df, pd.DataFrame(np.random.randn(10,5), columns=list('EDCBA'))], axis=1)

df.iloc[0, 2] = np.nan

df
```

```python
# Styling the table
# define colouring function
def colorNegValues(value):
    if value < 0:
        color = 'red'
    elif value > 0:
        color = 'black'
    else:
        color = 'green'
        
    return 'color: %s' % color

# apply the styling to the dataframe 
s = df.style.applymap(colorNegValues, subset=['A','B','C','D','E'])
s
```
